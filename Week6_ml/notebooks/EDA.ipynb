```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df_raw = pd.read_csv("../data/raw/titanic.csv")

df_raw.head()
df_raw.info()
df_raw.describe()
```

```python
df_raw.isnull().sum()

sns.heatmap(df_raw.isnull(), cbar=False)
plt.title("Missing Values (Raw Data)")
plt.show()
```

```python
df_raw.hist(figsize=(12,8))
plt.show()
```

```python
sns.countplot(x="Sex", hue="Survived", data=df_raw)
plt.show()

sns.countplot(x="Pclass", hue="Survived", data=df_raw)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df_raw.corr(numeric_only=True), annot=True)
plt.title("Correlation (Raw)")
plt.show()
```

```python
df = pd.read_csv("../data/processed/final.csv")

df.head()
df.info()
```

```python
sns.countplot(x="Survived", data=df)
plt.title("Target Distribution (Cleaned)")
plt.show()
```

```python
df.hist(figsize=(12,8))
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix (Cleaned)")
plt.show()
```
